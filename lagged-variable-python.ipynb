```python
import pandas as pd
shampoo = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/shampoo.csv')
shampoo.head()
```

```python
# Lag 1
shampoo['sales_lag_1'] = shampoo['Sales'].shift(1)
shampoo.head()
```

```python
# Lead 1
shampoo['sales_lead_1'] = shampoo['Sales'].shift(-1)
shampoo.head()
```
